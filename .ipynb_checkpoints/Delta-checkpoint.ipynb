```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings ('ignore')

%matplotlib inline
```

```python
df = pd.read_csv('objects/ES5min_04-24-2023.csv', parse_dates=True)
df['Date']=df['Date'] + " " + df['Time']
df['Date']=pd.to_datetime(df['Date'])
df.set_index(df['Date'], inplace = True)
df.dropna(inplace=True)
df.head()
```

```python
df.columns
```

```python
df = df.between_time(start_time='9:30:00', end_time='11:30:00', include_start='true', include_end='true' )
df = df[['Date', 'Time', 'Open', 'Last', 'High', 'Low', 'Volume', '# of Trades',
       'Bid Volume', 'Ask Volume', 'Delta']]
```

```python
df
```

```python
df30 = df.resample("30min")
df31 = df30.agg({
            "Open":  "first",
            "High":  "max",
            "Low":   "min",
            "Last": "last",
            "Volume": "sum",
            "Delta": "sum"
        })
locs = df31.index.indexer_at_time('9:30:00')
df32 = df31.iloc[locs]
df32.head(20)
df32.dropna(inplace=True)
df32
```

#### Does delta in the first 5 min indicate price action in subseqent periods. What is delta in the first 5 min period. What is the high of following 30min period?

```python
df['Delta_EMA-30'] = df['Delta'].rolling(60).mean()
```

```python
locs = df.index.indexer_at_time('9:30:00')
df = df.iloc[locs]
df.head(20)
df.dropna(inplace=True)
df['30min_high']=df32['High']
df['30min_low']=df32['Low']
df['sub_high_range'] = df['30min_high']-df['Last'] 
df['sub_low_range'] = df['30min_low']-df['Last']
df.columns
```

```python
d = df[['Delta', 'Delta_EMA-30', 'sub_high_range', 'sub_low_range',"Volume", '# of Trades']]
d['average_trades']=d['# of Trades'].rolling(30).mean()
d['average_volume']=d['Volume'].rolling(30).mean()
```

```python
# d =d.loc[(d['Delta']<d['Delta_EMA-30']) & (d['# of Trades']<d['average_trades']) & d['Volume']>d['average_volume']]
d =d.loc[(d['Delta']<d['Delta_EMA-30']) & (d['# of Trades']<d['average_trades'])]
d
```

```python
d['average']=(d['sub_high_range'] + d['sub_low_range']).rolling(60).mean()
```

```python
d['sub_low_range'].plot(figsize=(12,6))
d['sub_high_range'].plot()
d['average'].plot()
```
