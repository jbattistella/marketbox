```python
import pandas as pd
```

```python
df = pd.read_csv('objects/ES5min.csv', parse_dates=True)
df['Date']=df['Date'] + " " + df['Time']
df['Date']=pd.to_datetime(df['Date'])
df.set_index(df['Date'], inplace = True)
df.head()
```

```python
df30 = df.resample("30min")
df30.agg({
            "Open":  "first",
            "High":  "max",
            "Low":   "min",
            "Last": "last",
            "Volume": "sum"
        })
```
