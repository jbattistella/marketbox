```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df = pd.read_csv('objects/ES5min_04-24-2023.csv', parse_dates=True)
df['Date']=df['Date'] + " " + df['Time']
df['Date']=pd.to_datetime(df['Date'])
df.set_index(df['Date'], inplace = True)
df['Delta']
```

```python
df30 = df.resample("30min")
df31 = df30.agg({
            "Open":  "first",
            "High":  "max",
            "Low":   "min",
            "Last": "last",
            "Volume": "sum"
        })
locs = df31.index.indexer_at_time('9:30:00')
df32 = df31.iloc[locs]
df32.head(20)
df32.dropna(inplace=True)
df32['30m-range']=df32['High']-df32['Low']
df32['EMA-120'] = df32['30m-range'].ewm(alpha=1 / 6).mean()
df32
```

```python
df32['30m-range']['2023-01-01':].plot(figsize=(12,6),
                       legend=True)
df32['EMA-120']['2023-01-01':].plot(legend=True)
```

```python
df32['Volume_pct_chng']=(df32["Volume"]-df32["Volume"].shift(-1))/df32['Volume'].shift(-1)
df32['30m-range_pct_chng']=(df32['30m-range']-df32["30m-range"].shift(-1))/df32['30m-range'].shift(-1)
```

```python
df32['Volume_pct_chng'][240:].plot(figsize=(12,6), legend=True)
df32['30m-range_pct_chng'][240:].plot(legend=True)
```
